Python Project : Lead Scoring

Problem Statement:   

We have an online website which sells online courses to industry professionals. On any given day, many professionals who are interested in the courses land on their website and browse for the courses. Once these people land on the website, they might browse the courses or fill up a form for the course or watch some videos.


What are leads?

So, when these people fill up a form providing their email address or phone number, they are classified to be a lead.


What happens when the leads are received?

Once these leads are acquired or received, employees from the sales team start making calls, writing emails, etc. Through this process, some of the leads get converted while most do not. 


What is the lead conversion?

So, it’s basically a way/process of converting the leads received into paying customers.


Goal: Gather insights from the data, and find ways to make better decisions which would be helpful in growing the business.





Information about the dataset:


Variables Description


Customer ID - A unique ID with which the customer is identified.

Lead Source - The source of the lead. Includes Google, Organic Search, Olark Chat, etc.

Last Activity - Last activity performed by the customer. Includes Email Opened, Olark Chat Conversation, etc.

Lead Number - A lead number assigned to each lead obtained.

Do Not Email -An indicator variable selected by the customer wherein they select whether or not they want to be emailed about the course or not.

Do Not Call - An indicator variable selected by the customer wherein they select whether or not they want to be called about the course or not.

Lead Origin - The origin identifier with which the customer was identified to be a lead. Includes API, Landing Page Submission, etc.

Converted - The target variable. Indicates whether a lead has been successfully converted or not.

TotalVisits - The total number of visits made by the customer on the website.

Total Time Spent on Website - The total time spent by the customer on the website.

Page Views Per Visit - Average number of pages on the website viewed during the visits.

Country - The country of the customer.

City - The city of the customer.

Specialization - The industry domain in which the customer worked before. Includes the level 'Select Specialization' which means the customer had not selected this option while filling the form.

Sources - The source from which the customer heard about the company.

Current occupation - Indicates whether the customer is a student, unemployed or employed.

Reason - indicating what is their main motto behind doing this course.

Search - Indicating whether the customer had seen the ad in any of the listed items.

Magazine

Digital Advertisement

Through Recommendations - Indicates whether the customer came in through recommendations.

Receive More Updates - Indicates whether the customer chose to receive more updates about the courses.

Tags - Tags assigned to customers indicating the current status of the lead.

Lead Quality - Indicates the quality of lead based on the data and intuition of the employee who has been assigned to the lead.

Lead Profile - A lead level assigned to each customer based on their profile.

Pay-Cheque - Indicates whether the customer has agreed to pay the amount through cheque or not.

Free Demo - Indicates whether the customer wants a free copy of 'Mastering the Interview' or not.

Last Notable Activity - The last notable activity performed by the student.

```python
#Q1.Find the number of rows and columns in the dataset

print("Number of columns and rows are: ", data.shape)

rows = len(data.axes[0])
print("Number of Rows: ", rows)

cols = len(data.axes[1])
print("Number of Columns: ", cols)
```

```python
#Q2.(A.). Find out the number of customers who are outside of India

a = len(data.loc[data['Country'] != "India"])

print("The number of customers outside of India is = ", a)
```

```python
#Q2.(B.). Find out the number of customers from various countries and plot a graph for the same

print("Total number of countries: ", len(pd.unique(data['Country'])))

CountOfCustomersFromVariousCountries = data.groupby('Country')['Customer ID'].nunique()
print(CountOfCustomersFromVariousCountries)

fig = plt.figure(figsize =(100,50))

plt.bar(CountOfCustomersFromVariousCountries.index, CountOfCustomersFromVariousCountries.values)

plt.xlabel('Countries', fontweight ='bold', fontsize = 100)
plt.ylabel('No. of Customers', fontsize = 150)
plt.title("Customers from various countries",fontweight ='bold', fontsize = 100)
plt.show()
```

```python
#Q2.(C.). Find out which country has the highest number of customers

CountryWithTheHighestNoOfCustomers = data.groupby('Country')['Customer ID'].nunique().idxmax()
print("The country with the highest number of customers is: ", CountryWithTheHighestNoOfCustomers)
```

```python
#Q3.Create a new column as index in the dataset

data['Index'] = data.index
data.head(10)
```

```python
#Q4. Using loop change the values of ‘Select’ ‘NaN’ of city column

for i in data.iterrows():
    if i[1]['City'] == 'Select':
        data['City'][i[0]] == np.nan
        
        
data['City'].value_counts()
```

```python
#Q5. Replace all the missing/blank values in the dataset with NaN

data = data.replace(' ', np.nan, regex = True)
data.head(10)
```

```python
#Q6. Find the average time spent on the website

AverageSecondsSpentOnTheWebsite = data["Total Time Spent on Website"].mean()
print("The average seconds spent on the website is: ", AverageSecondsSpentOnTheWebsite)
```

```python
#Q7. The column ‘total time spent on the website’ is in seconds. Convert that to a desired form(min or hours) 
#and find the highest count and plot an applicable graph for the same.

import datetime

data['Total Minutes Spent on Website'] = pd.to_timedelta(data['Total Time Spent on Website'], unit='second')
print(data['Total Minutes Spent on Website'])
```

```python
print("The highest time spent on the website is: ", max(data['Total Minutes Spent on Website']))
```

```python
#Q8. Convert the column of ‘Converted’ from values 1/0 to yes/no and find the count of yes and no and plot a graph for the same.

data.Converted.replace([0.0, 1.0],['no','yes'], inplace=True)

print(data.Converted.value_counts())

plt.bar(data.Converted.value_counts().index, data.Converted.value_counts().values)
plt.show()
```

```python
#Q9. Check how the total time spent on the website and converted are related.

data.Converted.replace(['no','yes'], [0.0, 1.0], inplace=True)

print(data.Converted.value_counts())
```

```python
df = data[['Total Time Spent on Website', 'Converted']]
print(df.head(15))

corr = df['Total Time Spent on Website'].corr(df['Converted'])
print("The correlation between 'Total Time Spent on Website' and 'Converted' is: ", corr)
```
